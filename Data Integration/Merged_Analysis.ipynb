```python
import pandas as pd
```

```python
enrollment_df = pd.read_csv(r"D:\UIDAI\aadhar_enrollment_cleaned(1).csv")
demographic_df = pd.read_csv(r"D:\UIDAI\aadhar_demographic_cleaned(1).csv")
biometric_df = pd.read_csv(r"D:\UIDAI\aadhar_biometric_cleaned(1).csv")
```

```python
#Verify Structure Before Merging
enrollment_df.info()
demographic_df.info()
biometric_df.info()
```

```python
#Define Common Merge Keys
common_keys = ['date', 'state', 'district', 'pincode', 'month']
```

```python
#Merge Enrollment + Demographic Data
merged_1 = pd.merge(
    enrollment_df,
    demographic_df,
    on=common_keys,
    how='inner'
)
```

```python
#Merge the Result with Biometric Data
final_merged_df = pd.merge(
    merged_1,
    biometric_df,
    on=common_keys,
    how='inner'
)

```
